```python
import pandas as pd
import cv2
import numpy as np
from PIL import Image
import os
import shutil
```

```python
dataset_path = 'fer2013/fer2013.csv'
image_size=(48,48)
```

```python
data = pd.read_csv(dataset_path)
```

```python
data.head()
```

```python
def make_names(x):
    if x==0:
        return 'Angry'
    if x==1:
        return 'Disgust'
    if x==2:
        return 'Fear'
    if x==3:
        return 'Happy'
    if x==4:
        return 'Sad'
    if x==5:
        return 'Surprise'
    if x==6:
        return 'Neutral'
    
```

```python
data['emotion']=data['emotion'].apply(lambda x: make_names(x))
```

```python
def load_fer2013(data):
    pixels = data['pixels'].tolist()
    width, height = 48, 48
    for z,pixel_sequence in enumerate(pixels):
        data_=data[data['pixels']==pixel_sequence]
        face = [int(pixel) for pixel in pixel_sequence.split(' ')]
        face = np.asarray(face).reshape(width, height)
        face = cv2.resize(face.astype('uint8'),image_size)
        img = Image.fromarray(face.astype('float32')).convert('RGB')
        emotion=data_['emotion'].values[0]
        img.save('all_the_emotions/{}{}.jpg'.format(emotion,z))
    
 
```

```python
def make_dirs_and_put_data(data):
    emotions=data['emotion'].unique()
    for i in emotions:
        if not os.path.exists('all_the_emotions/{}'.format(i)):
           os.mkdir('all_the_emotions/{}'.format(i))
    for i in os.listdir('all_the_emotions'):
        
        if not os.path.isdir('all_the_emotions/'+i):
            for z in emotions:
                if z in i:
                    print(i)
                    shutil.move('all_the_emotions/{}'.format(i),'all_the_emotions/{}/{}'.format(z,i))
        
    
```

```python
load_fer2013(data)
```

```python
make_dirs_and_put_data(data)
```
